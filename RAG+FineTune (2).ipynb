```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from peft import PeftModel
import os

# =========================
# CONFIG
# =========================

BASE_MODEL = "mistralai/Mistral-7B-Instruct-v0.2"
ADAPTER_PATH = "/content/drive/MyDrive/mistral-7b-finetuned-lora"

device = "cuda" if torch.cuda.is_available() else "cpu"

# =========================
# VERIFY PATH (IMPORTANT DEBUG STEP)
# =========================

print("Files in adapter path:", os.listdir(ADAPTER_PATH))

# =========================
# TOKENIZER
# =========================

tokenizer = AutoTokenizer.from_pretrained(BASE_MODEL)
tokenizer.pad_token = tokenizer.eos_token

# =========================
# BASE MODEL (4-bit)
# =========================

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16
)

base_model = AutoModelForCausalLM.from_pretrained(
    BASE_MODEL,
    quantization_config=bnb_config,
    device_map="auto"
)

# =========================
# LOAD ADAPTER (THIS WILL NOW WORK)
# =========================

model = PeftModel.from_pretrained(
    base_model,
    ADAPTER_PATH,
    local_files_only=True   # 🔥 KEY FIX
)

model.eval()

print("✅ MODEL + ADAPTER LOADED SUCCESSFULLY")
```

```python
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings

persist_dir = "/content/drive/MyDrive/IITB_VectorDB"

embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2",
    model_kwargs={"device": "cuda"}
)

vectorstore = Chroma(
    persist_directory=persist_dir,
    embedding_function=embedding_model
)

retriever = vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 2}
)

print("Vector DB loaded")
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
import torch

model_name = "mistralai/Mistral-7B-Instruct-v0.2"

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4"
)

tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "left"

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    quantization_config=bnb_config
)

print("Model loaded")
```

```python
def rag_query(question):
    docs = retriever.invoke(question)

    context = "\n\n".join([doc.page_content[:300] for doc in docs])

    prompt = f"""<s>[INST]
Answer using ONLY the context below.

Context:
{context}

Question:
{question}
[/INST]
"""

    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=100,
            do_sample=False,
            pad_token_id=tokenizer.eos_token_id
        )

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return response.split("[/INST]")[-1].strip()
```

```python
print(rag_query("What is IIT Bombay?"))
print("--------------------------------------------------")
print(rag_query("What programs are offered at IIT Bombay?"))
```
