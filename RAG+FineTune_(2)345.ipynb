```python
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings

persist_dir = "/content/drive/MyDrive/IITB_VectorDB2"

embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2",
    model_kwargs={"device": "cuda"}
)

vectorstore = Chroma(
    persist_directory=persist_dir,
    embedding_function=embedding_model
)

retriever = vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 5}
)

print("Vector DB loaded")
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
import torch

model_name = "mistralai/Mistral-7B-Instruct-v0.2"

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4"
)

tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "left"

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    quantization_config=bnb_config
)

print("Model loaded")
```

```python
def rag_query(question):
    # 1. Retrieve relevant documents
    docs = retriever.invoke(question)

    # 2. Prepare Context: Removed the [:300] slice to ensure full information is passed
    context = "\n\n".join([doc.page_content for doc in docs])

    # 3. Enhanced "Digital Senior" Prompt Template
    prompt = f"""<s>[INST]
<<SYS>>
You are the "Digital Senior," the official AI assistant for IIT Bombay.
Your goal is to help students (especially freshers) find accurate information regarding
academic rules, hostel regulations, and campus life based ONLY on the official context provided below.
<</SYS>>

### INSTRUCTIONS:
- Answer ONLY using the provided Context. Do not use external knowledge.
- If the answer is not in the context, say: "I'm sorry, that specific information isn't in the official handbooks I have access to. Please check the IITB Academic Office or your Hostel Warden."
- Use bullet points for multi-step rules to make them easy to read.
- Maintain a helpful and professional tone.

### OFFICIAL CONTEXT:
{context}

### STUDENT QUESTION:
{question}
[/INST]
"""

    # 4. Tokenization
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

    # 5. Generation with Precision Settings
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=200,      # Increased slightly for detailed rules
            do_sample=True,          # Enabled sampling for natural flow
            temperature=0.1,         # Very low temp keeps it strictly factual
            repetition_penalty=1.1,  # Prevents the model from getting stuck in loops
            pad_token_id=tokenizer.eos_token_id
        )

    # 6. Clean and Return Output
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response.split("[/INST]")[-1].strip()
```

```python
print(rag_query("What is IIT Bombay?"))
print("--------------------------------------------------")
print(rag_query("What programs are offered at IIT Bombay?"))
```

```python
from google.colab import drive
drive.mount('/content/drive')

import json
import numpy as np
from sentence_transformers import SentenceTransformer, util

# Load embedding model
embed_model = SentenceTransformer("all-MiniLM-L6-v2")

# 🔁 UPDATE THIS PATH if your file is inside a folder
json_path = "/content/drive/MyDrive/test2.json"

# Load test data
with open(json_path, "r") as f:
    test_data = json.load(f)

def evaluate_system():
    answer_sim_scores = []
    retrieval_sim_scores = []

    for item in test_data:
        question = item["question"]
        ground_truth = item["answer"]

        # Generate answer
        generated_answer = rag_query(question)

        # ===== Answer Similarity =====
        emb_pred = embed_model.encode(generated_answer, convert_to_tensor=True)
        emb_gt = embed_model.encode(ground_truth, convert_to_tensor=True)

        sim_score = util.cos_sim(emb_pred, emb_gt).item()
        answer_sim_scores.append(sim_score)

        # ===== Retrieval Evaluation =====
        docs = retriever.invoke(question)

        q_emb = embed_model.encode(question, convert_to_tensor=True)

        sims = []
        for d in docs:
            c_emb = embed_model.encode(d.page_content, convert_to_tensor=True)
            sims.append(util.cos_sim(q_emb, c_emb).item())

        avg_retrieval_sim = np.mean(sims)
        retrieval_sim_scores.append(avg_retrieval_sim)

        # Print results per question
        print("\n-----------------------------")
        print("Question:", question)
        print("Generated:", generated_answer)
        print("Ground Truth:", ground_truth)
        print("Answer Similarity:", round(sim_score, 3))
        print("Retrieval Avg Similarity:", round(avg_retrieval_sim, 3))

    # Final results
    print("\n=============================")
    print("FINAL RESULTS")
    print("=============================")
    print("Average Answer Similarity:", np.mean(answer_sim_scores))
    print("Average Retrieval Similarity:", np.mean(retrieval_sim_scores))


# ▶ Run evaluation
evaluate_system()
```
